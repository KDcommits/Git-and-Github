## STEP 3 : Untracked and Staged files


- Staging helps to choose what files you would like to push for commit. Say there are 4 files we have worked on but we want to push only 2 files so before commit we would stage the files so that we can commit and push only those 2. Commiting directly will lead to make changes in all 4 files.

- Initialize the git repository first if you are creating a repo from the scratch.Command :
   - ***git init***
- After initializing the repository a file with .git extension will be created which will contain the binaries.
- Create a dummy file main.py in the same folder
- To know the status of the files that are pushed via commit
    - ***git status***
-  The below ss shows the files that we have currently are untracked.

- Now we need to bring those files into the staging area that we would like to push for commit. Command to bring files into staging area:
    - ***git add \<filename\>***
    - ***git add -A***     <i>---- Adds all the files within the directory to the staging area </i>
- Now when we again check for status then it says "no commits yet". So we need to make a commit for the staged files.


- Now run the command ***git commit***. It will commit the changes only for the staged files and open vm editor. Press "i" key on the keyboard to write a text. This is basically the message for that commit so that other developers can understand what the commit is for. <br>


-  When you are done with writing the commit message, then click on "**esc**" key and then write **:wq** to save the commit. It will exit the vm editor and the commit will be shown.
- The below screenshot shows the status of the file "Git and Github.ipynb" is untracked. This is because we had only staged for "main.py" file. This proves the commit that we did only made changes to the "main.py" file.

- There is a simpler way of pushing commits without even opening vm editor.Command:
    - ***git commit -m "\<type the message\>"***

### NOTES:

- Command to get the last committed version of a file : ***git checkout \<filename\>*** 
- Command to get the last committed version of all files : ***git checkout -f*** 
- Command to check all the commit history : ***git log***
- Command to check only the last n commit hostory : ***git log -p -\<n\>***
